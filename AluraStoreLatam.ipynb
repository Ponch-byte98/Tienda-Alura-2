```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head(20)
```

#1. Análisis de facturación



```python
#Código para la suma de ingresos por tienda
ingreso_tienda1 = tienda['Precio'].sum()
ingreso_tienda2 = tienda2['Precio'].sum()
ingreso_tienda3 = tienda3['Precio'].sum()
ingreso_tienda4 = tienda4['Precio'].sum()

def formato_pesos(valor):
  return f'$ {valor:,.2f}'

print("Ingresos totales por tienda")
print(f'Tienda 1: {formato_pesos(ingreso_tienda1)}')
print(f'Tienda 2: {formato_pesos(ingreso_tienda2)}')
print(f'Tienda 3: {formato_pesos(ingreso_tienda3)}')
print(f'Tienda 4: {formato_pesos(ingreso_tienda4)}')
```

```python
import matplotlib.pyplot as plt

ingresos = [tienda['Precio'].sum(),
            tienda2['Precio'].sum(),
            tienda3['Precio'].sum(),
            tienda4['Precio'].sum()
            ]
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

#Crear gráfico
plt.bar(tiendas,ingresos)
# Añadir título y etiquetas a los ejes
plt.title('Gráfico de Barras de ingresos')
plt.xlabel('Ingresos')
plt.ylabel('Tiendas')
plt.ticklabel_format(style='plain', axis='y') #Cambiar los valores de la escala del eje de las X

# Mostrar el gráfico
plt.show()
```

# 2. Ventas por categoría

```python
def categorias(tienda_df, nombre_tienda):
    top_categorias = tienda_df['Categoría del Producto'].value_counts().head(3)
    print(f"\nLos 3 productos mas vendidos por categoría de la {nombre_tienda}:")
    for categoria, cantidad in top_categorias.items():
        print(f"- {categoria}: {cantidad} ventas")

categorias(tienda, "Tienda 1")
categorias(tienda2, "Tienda 2")
categorias(tienda3, "Tienda 3")
categorias(tienda4, "Tienda 4")
```

```python
#Gráfica de ventas por categoría
def grafica_ventas(tienda_df, nombre_tienda):
    top_categorias = tienda_df['Categoría del Producto'].value_counts().head(3)
    plt.figure(figsize=(8, 5))
    top_categorias.plot(kind='bar', color='blue')
    plt.title(f'Los 3 productos más vendidos por categoría de la {nombre_tienda}')
    plt.ylabel('Cantidad de Ventas')
    plt.xlabel('Categorías')
    plt.xticks(rotation=0)
    plt.grid(axis='y', linestyle='--', alpha=0.5)
    plt.tight_layout()
    plt.show()
grafica_ventas(tienda, "Tienda 1")
grafica_ventas(tienda2, "Tienda 2")
grafica_ventas(tienda3, "Tienda 3")
grafica_ventas(tienda4, "Tienda 4")
```

# 3. Calificación promedio de la tienda


```python
calificacion_tienda1 = tienda['Calificación'].mean()
calificacion_tienda2 = tienda2['Calificación'].mean()
calificacion_tienda3 = tienda3['Calificación'].mean()
calificacion_tienda4 = tienda4['Calificación'].mean()

print('Calificación promedio por tienda:')

print(f"Calificación promedio Tienda 1: {calificacion_tienda1:.2f}")
print(f"Calificación promedio Tienda 2: {calificacion_tienda2:.2f}")
print(f"Calificación promedio Tienda 3: {calificacion_tienda3:.2f}")
print(f"Calificación promedio Tienda 4: {calificacion_tienda4:.2f}")
```

```python
calificaciones_promedio = [tienda['Calificación'].mean(),
                           tienda2['Calificación'].mean(),
                           tienda3['Calificación'].mean(),
                           tienda4['Calificación'].mean()
                           ]
nombre_tiendas = ['Tienda 1','Tienda 2','Tienda 3','Tienda 4']
colores = ["#EE6055","#60D394","#AAF683","#FFD97D","#FF9B85"]
plt.pie(calificaciones_promedio, labels=nombre_tiendas, autopct="%0.1f %%", colors=colores)
plt.axis("equal")
plt.show()
```

# 4. Productos más y menos vendidos

```python
def analizar_ventas_por_producto(df, nombre_tienda):
    ventas = df['Producto'].value_counts()
    producto_mas_vendido = ventas.idxmax()
    producto_menos_vendido = ventas.idxmin()

    print(f"{nombre_tienda}")
    print(f"Más vendido: {producto_mas_vendido} ({ventas.max()} ventas)")
    print(f"Menos vendido: {producto_menos_vendido} ({ventas.min()} venta(s))")


analizar_ventas_por_producto(tienda, 'Tienda 1')
analizar_ventas_por_producto(tienda2, 'Tienda 2')
analizar_ventas_por_producto(tienda3, 'Tienda 3')
analizar_ventas_por_producto(tienda4, 'Tienda 4')
```

```python
def graficar_ventas_por_producto_pie(df, nombre_tienda):
    ventas = df['Producto'].value_counts()
    producto_mas_vendido = ventas.idxmax()
    producto_menos_vendido = ventas.idxmin()
    cantidad_mas = ventas.max()
    cantidad_menos = ventas.min()

    labels = [f'Más vendido: {producto_mas_vendido}', f'Menos vendido: {producto_menos_vendido}']
    sizes = [cantidad_mas, cantidad_menos]
    colors = ['blue', 'skyblue']
    plt.figure(figsize=(6, 6))
    plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
    plt.title(f'Comparación de Ventas en {nombre_tienda}')
    plt.axis('equal')  # Círculo perfecto
    plt.tight_layout()
    plt.show()

# Ejecutar para cada tienda
graficar_ventas_por_producto_pie(tienda, 'Tienda 1')
graficar_ventas_por_producto_pie(tienda2, 'Tienda 2')
graficar_ventas_por_producto_pie(tienda3, 'Tienda 3')
graficar_ventas_por_producto_pie(tienda4, 'Tienda 4')
```

# 5. Envío promedio por tienda

```python
costo_envio_tienda1 = tienda['Costo de envío'].mean()
costo_envio_tienda2 = tienda2['Costo de envío'].mean()
costo_envio_tienda3 = tienda3['Costo de envío'].mean()
costo_envio_tienda4 = tienda4['Costo de envío'].mean()

def corregir_formato(valor):
  return f'$ {valor:,.2f}'

print("Costo promedio de envio por tienda")
print(f'Tienda 1: {corregir_formato(costo_envio_tienda1)}')
print(f'Tienda 2: {corregir_formato(costo_envio_tienda2)}')
print(f'Tienda 3: {corregir_formato(costo_envio_tienda3)}')
print(f'Tienda 4: {corregir_formato(costo_envio_tienda4)}')
```

```python
envio_promedio = [tienda['Costo de envío'].mean(),
                  tienda2['Costo de envío'].mean(),
                  tienda3['Costo de envío'].mean(),
                  tienda4['Costo de envío'].mean()
            ]

tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

#Crear gráfico
plt.bar(tiendas,envio_promedio)
# Añadir título y etiquetas a los ejes
plt.title('Gráfico de Barras de envio promedio por tienda')
plt.xlabel('Tiendas')
plt.ylabel('Envio promedio')
plt.ticklabel_format(style='plain', axis='y') #Cambiar los valores de la escala del eje de las X


# Mostrar el gráfico
plt.show()
```
